```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

Import The Data 

```python
df = pd.read_csv("olympics.csv")
df
```

```python
df.rename(columns={ '01 !' : 'sGold' , '02 !' : 'sSilver' , '03 !' : 'sBronze'},inplace = True)
df.rename(columns={ '01 !.1' : 'wGold' , '02 !.1' : 'wSilver' , '03 !.1' : 'wBronze'},inplace = True)
df.rename(columns={ '01 !.2' : 'Gold' , '02 !.2' : 'Silver' , '03 !.2' : 'Bronze'},inplace = True)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe(include = "all")
```

```python
max_win = df['Combined total'].max()
max_name = df.loc[df['Combined total'] == max_win,'Countries'].values[0]
print(f" The {max_name} won maximum numbers of medals {max_win}")
```

```python
top_wins = df.sort_values(by = 'Combined total',ascending = False).head(10)
top_names = top_wins[['Countries','Combined total']]
top_names
```

```python
plt.figure(figsize = (10,8))
ax = sns.barplot(x = 'Countries' ,y = 'Combined total', data = top_names, palette = 'husl')
plt.xticks(rotation = 30,ha = 'right')
for p in ax.patches:
    ax.annotate(format(p.get_height()),
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='center', xytext=(0, 9), textcoords='offset points') 
plt.show()
```

```python
summer_top = df.sort_values(by = 'Total', ascending = False).head(10)
top = summer_top[['Countries','Total']]
top
```

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(x = 'Countries' , y  = 'Total' , data = top , palette = 'tab10')
plt.title("Top 10 Summer Winners")
plt.xlabel("Countires")
plt.ylabel("Numbers of Medal wins")
plt.xticks(rotation = 45 ,ha = 'right')
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'),
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 9), textcoords='offset points')  
plt.show()
```

```python
winter_wins = df.sort_values(by = 'Total.1',ascending = False).head(10)
top_winters = winter_wins[['Countries','Total.1']]
top_winters
```

```python
plt.figure(figsize = (10,8))
ax = sns.barplot(x = 'Countries' , y = 'Total.1' , data = top_winters,palette = 'vlag')
plt.title("Top 10 Winter Winners")
plt.xlabel("Counteris")
plt.ylabel("Numbers of winners")
plt.xticks(rotation = 45 , ha = 'right') 
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'),
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 9), textcoords='offset points')  
plt.show()
```
